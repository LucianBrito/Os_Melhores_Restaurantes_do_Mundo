```python
# Importar bibliotecas necessárias
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotar distribuição dos dados de todas as variaveis numéricas
df.hist(bins=15, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Plotar distribuição dos dados das variaveis, cidade, Pais,Preço por categoria, Continente, Categoria de estrelas
categorical_columns = ['Country', 'Price_Category', 'Continent', 'Star_Category']
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)
    plt.title(f'Distribution of {column}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
df.info()
```

```python
# Analise de Restaurantes por Pais
country_counts = df['Country'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=country_counts.index, y=country_counts.values, palette='viridis')
plt.title('Number of Top Restaurants by Country')
plt.xlabel('Country')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Restaurantes por Continente
continent_counts = df['Continent'].value_counts()
plt.figure(figsize=(10, 5)) 
sns.barplot(x=continent_counts.index, y=continent_counts.values, palette='magma')
plt.title('Number of Top Restaurants by Continent')
plt.xlabel('Continent')
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
# Quantidade de Restaurantes por numero de estrelas
star_counts = df['Star_Category'].value_counts().sort_index()
plt.figure(figsize=(8, 5))
sns.barplot(x=star_counts.index, y=star_counts.values, palette='coolwarm')
plt.title('Number of Top Restaurants by Star Category')
plt.xlabel('Star Category')
plt.ylabel('Number of Restaurants')
plt.tight_layout()
plt.show()
```

```python
# Qual a relação entre estrelas e Avaliação?
plt.figure(figsize=(10, 6))
sns.boxplot(x='Star_Category', y='Rating', data=df, palette='Set2')
plt.title('Rating Distribution by Star Category')
plt.xlabel('Star Category')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
# corralação entre Price_Category x Rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price_Category', y='Rating', data=df, hue='Continent', palette='Set1')
plt.title('Price Category vs Rating by Continent')

```

```python
# Correlação de Estrela x Preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='Star_Category', y='Price_Category', data=df, palette='Set3')
plt.title('Price Category Distribution by Star Category')   
```

```python
# Quais chefes aparecem mais vezes na lista dos melhores restaurantes do mundo?
top_chefs = df['Chef'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_chefs.index, y=top_chefs.values, palette='pastel')
plt.title('Top 10 Chefs with Most Restaurants in the List')
plt.xlabel('Chef')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
